```python
'''
Assignment Tasks:
1. Reading the CSV File:
- Load the provided CSV file into a pandas DataFrame.
- Display the first 5 rows of the DataFrame to understand the structure of the data.
2. Basic Analysis:
- Calculate the total points scored by all players combined.
- Find the player who has played the most minutes (MP).
- Identify the top 5 players in terms of total rebounds (TRB).
3. Player Filtering:
- Create a new DataFrame containing only players who played more than 500 minutes.
- From this filtered DataFrame, determine the player with the highest total assists (AST).
- Who are the top 3 Assist Leads in the League?
- Who are the top 3 Shot Blockers?
4. School-Based Analysis:
- Group the players by School and calculate the total points scored by each school.
- Group the players by School and calculate the total assists (AST) for each team.
- Sort the schools by total points scored and display the top 3 schools.
5. Extra Credit (Bonus):

'''
```

```python
import pandas as pd

bb_df = pd.read_csv('acc_players-2324F.csv', header = 1)

bb_df.head(5)
```

```python
import numpy as np

bb_df.info()

print('Total Points Scored by all Players:', np.sum(bb_df['PTS']), 'points')

print('Player Who Scored Most Points:')

display(bb_df[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1))


print('Top 5 Rebound Players:')
bb_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).groupby('Player').head(5)
```

```python
'''
3. Player Filtering:
- Create a new DataFrame containing only players who played more than 500 minutes.
- From this filtered DataFrame, determine the player with the highest total assists (AST).
- Who are the top 3 Assist Leads in the League?
- Who are the top 3 Shot Blockers?
'''
gr500_df = bb_df[bb_df['MP']>500]
print('Player with Highest Total Assists:')
display(gr500_df[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(1))

print('Top 3 Assist Leads in the League:')
display(gr500_df[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(3))

print('Top 3 Shot Blockers in the League:')
display(gr500_df[['Player', 'BLK']].sort_values(by = 'BLK', ascending = False).head(3))
```

```python
'''
4. School-Based Analysis:
- Group the players by School and calculate the total points scored by each school.
- Group the players by School and calculate the total assists (AST) for each team.
- Sort the schools by total points scored and display the top 3 schools.
'''

print('Total Points Scored by Each School:')
display(bb_df.groupby('School')[['School', 'PTS']].sum('PTS'))

print('Total Assists Scored by Each School:')
display(bb_df.groupby('School')[['School', 'AST']].sum('AST'))

print('Total Points Scored by Top 3 Schools:')
display(bb_df.groupby('School')[['School', 'PTS']].sum('PTS').head(3))
```
