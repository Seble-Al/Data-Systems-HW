```python
import numpy as np
import pandas as pd
ghi_data = pd.read_csv('GHI_data.csv', sep=',', encoding='latin1', header = 0)
df_ghi = ghi_data.rename(columns={'ï»¿Country': 'Country', '2008 GHI\n': '2008 GHI'})
df_ghi = df_ghi.drop([136,137,138,139], axis='index')
df_ghi
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_ghi["% GHI change since 2016"] = pd.to_numeric(df_ghi["% GHI change since 2016"])

df_ghi.dtypes
```

```python
numeric_ghi = df_ghi.select_dtypes(include=['float64'])
correlation_matrix = numeric_ghi.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('GHI Heatmap of Numeric Features in GHI Dataset')
plt.show()
```
